# Evaluación de Golden Set con MLflow

Este notebook ejecuta la evaluación de Task1 y Task2 sobre el golden set, 
loggea métricas a MLflow y reporta F1/Accuracy/Recall/Precision y tiempo de ejecución.


```python
from pathlib import Path
import sys
import pandas as pd

# Importar script de evaluación
sys.path.append(str(Path.cwd() / 'evaluation' / 'scripts'))
import eval_gold as eg

# Encoder igual al usado en la app (BETO/RoBERTa-BNE)
ENCODER_VARIANT = 'roberta'
```

```python
# Verificar tamaños de los datasets
t1 = pd.read_csv('data_lake/datasets/task1_gold_labeled.csv')
t2 = pd.read_csv('data_lake/datasets/task2_gold_labeled.csv')
print('Task1 rows:', len(t1))
print('Task2 rows:', len(t2))
```

```python
# Task1 — batch grande (200) recomendado para LLM
res_t1_encoder = eg.evaluate_task1('encoder', encoder_variant=ENCODER_VARIANT, batch_size=200)
res_t1_llm = eg.evaluate_task1('llm', encoder_variant=ENCODER_VARIANT, batch_size=200)
res_t1_api = eg.evaluate_task1('api', encoder_variant=ENCODER_VARIANT, batch_size=200)
[res_t1_encoder, res_t1_llm, res_t1_api]
```

```python
# Task2 — batch moderado (100)
res_t2_encoder = eg.evaluate_task2('encoder', encoder_variant=ENCODER_VARIANT, batch_size=100)
res_t2_llm = eg.evaluate_task2('llm', encoder_variant=ENCODER_VARIANT, batch_size=100)
res_t2_api = eg.evaluate_task2('api', encoder_variant=ENCODER_VARIANT, batch_size=100)
[res_t2_encoder, res_t2_llm, res_t2_api]
```

```python
# Tabla resumen
results = [res_t1_encoder, res_t1_llm, res_t1_api, res_t2_encoder, res_t2_llm, res_t2_api]
df = pd.DataFrame([r.__dict__ for r in results])
df
```
